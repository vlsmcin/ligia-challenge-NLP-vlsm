# Inferência

Este notebook executa a inferencia no conjunto de teste final do Kaggle e gera o arquivo de submissao.
Ele demonstra a integracao do modelo treinado em um fluxo completo de previsao.

### Importando as bibliotecas

```python
from scipy.sparse import load_npz
import pandas as pd
import joblib
```

### Carregando os dados e o modelo

Para carregar o `X_final_test` ja preprocessado, execute antes o notebook 2 (Pre-Processamento).
Esse passo gera o arquivo `../data/processed/X_final_test.npz`.

```python
X_final_test = load_npz("../data/processed/X_final_test.npz")
model = joblib.load("../models/best_model.joblib")
test = pd.read_csv("../data/raw/test.csv")
```

### Fazendo inferência

```python
y_final_test_pred = model.predict(X_final_test)
submission = pd.DataFrame({
    'id': test['id'],
    'label': y_final_test_pred
})
submission.to_csv('../data/processed/submission.csv', index=False)
```
